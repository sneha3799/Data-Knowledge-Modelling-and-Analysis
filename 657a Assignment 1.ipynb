```python
# using interquartile range (IQR)
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

```python
outliers = find_outliers_IQR(df)
outliers.isna().sum()
```

```python
# trimming - drop the outliers
def drop_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    outliers_dropped = not_outliers.dropna()
    return not_outliers,outliers_dropped
```

```python
not_outliers, outliers_dropped = drop_outliers_IQR(df)
not_outliers.shape,outliers_dropped.shape
```

```python
# replace outliers using imputations as if they were missing values
def impute_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    upper = df[~(df>(q3+1.5*IQR))].max()
    lower = df[~(df<(q1-1.5*IQR))].min()
    df = np.where(df > upper,
                  df.mean(),
                  np.where(
                      df < lower,
                      df.mean(),
                      df
                  )
                 )
    return df
```

```python
df_new = impute_outliers_IQR(df)
df_new.shape
```
